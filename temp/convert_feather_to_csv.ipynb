```python
import os 
import pandas as pd

option_dir = r'C:\Users\kylec\OneDrive\Desktop\react_project\my-tutorial\vertiflow\public\data_dev\Direct-3Zone-DD-Lunch'
zones = [f for f in os.listdir(option_dir) if os.path.isdir(os.path.join(option_dir, f))]
for zone in zones:
    zone_dir = os.path.join(option_dir, zone)
    configs = [f for f in os.listdir(zone_dir) if os.path.isdir(os.path.join(zone_dir, f))]
    for config in configs:
        config_dir = os.path.join(zone_dir, config)
        runs = [f for f in os.listdir(config_dir) if os.path.isdir(os.path.join(config_dir, f))]
        for run in runs:
            run_dir = os.path.join(config_dir, run)
            feather_files = [f for f in os.listdir(run_dir) if f.endswith('.feather')]
            parquet_files = [f for f in os.listdir(run_dir) if f.endswith('.parquet')]
            for feather_file in feather_files:
                if feather_file.replace('.feather', '.parquet') in parquet_files: continue
                feather_path = os.path.join(run_dir, feather_file)
                df = pd.read_feather(feather_path)
                parquet_path = os.path.join(run_dir, feather_file.replace('.feather', '.parquet'))
                df.to_parquet(parquet_path, compression='snappy', index=False)
                print(f"Converted: {feather_file} for {zone} - {config} - {run}")
```
